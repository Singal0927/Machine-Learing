```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris=sns.load_dataset('iris')
iris.head(5)
```

```python
iris.sepal_length.hist(bins=20,color='y')#bins代表数据分组
```

```python
iris.hist()
```

```python
iris.plot.density(subplots=True)#核密度图，subplots=True代表分图显示
```

```python
sns.distplot(iris['sepal_length'],rug=True)#rug表示地毯图
```

```python
iris.plot.box()
```

```python
sns.boxplot(x='species',y='sepal_width',data=iris)#分类画箱型图
```

```python
sns.scatterplot(x='petal_width',y='petal_length',data=iris,hue='species',style='species')#hue表示上色，style表示图形样式
```

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris,diag_kind='kde',vars=['sepal_width','sepal_length','petal_width'],hue='species')#diag_kind表示对角线图的种类
```

```python
sns.jointplot(x='sepal_width',y='sepal_length',data=iris,kind='kde')#二维核密度图
```
